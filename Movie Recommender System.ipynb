```python
# Need to convert List of  Dictionary 
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# to '{'Action,'Adventure','FFantasy','SciFi'}'
```

```python
#convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
# So we are actually passing the String not the list of dictionaries so first we need to convert this.
```

```python
# String to List
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
#Helper Function to achive the above
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
columns
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

Keywords Column

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
columns
```

```python
movies.iloc[0].cast
```

```python
#Helper Function to select only top 3 cast
def convert_cast(obj):
    L=[]
    number=0
    for i in ast.literal_eval(obj):
        number+=1
        if number>3:
            break
        else:
            L.append(i['name'])
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

Crew Column

```python
ast.literal_eval(movies['crew'][0])
```

```python
#Helper Function to achive the director of the movie from the crew members
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if(i['job'] == 'Director'):
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

# Text Vectorization - Bag of Words

Steming Method

```python
#['loved','loving','love'] --> ['love','love','love']
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# Helper Function
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'][0]
```

```python
stem(new_df['tags'][0])
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words ='english')
```

```python
# every movie is converted in to vector
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
# To get know what are the most top 5000 words which got repeated
cv.get_feature_names_out()
```

```python
vectors.shape
```

```python
# Now to find the similarity between the movies, distance between the vectors(movies) is being calculated.
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# It gives an array of arrays of shape (4806,4806)
similarity = cosine_similarity(vectors)
```

```python
# It shows the similarity of the first movie with the all the other 4806 movies and its an array
similarity[0]
```

# Function for Recommending Movies

1. Find the index of the movie in the similarity matrix.
2. After locating the movie in the similarity array then sort the array, because each element of the array measures the similarity of the present movie with the other movies.
3. From that then select the top 5 Movies.

```python
# Step 1: method for finding the index of the film from the new_df
new_df[new_df['title'] == 'Batman Begins'].index[0]
```

Cosine_similarity

This measures how similar two vectors are.

Range: -1 to 1 (but usually between 0 and 1 for most use cases like BoW where values are non-negative).

1 means: exactly the same direction (high similarity)

0 means: completely different (no similarity)

Cosine_distances

This measures how dissimilar two vectors are.

Formula: 1 - cosine_similarity

Range: 0 (very similar) to 2 (completely opposite)

For most normal use cases, values fall between 0 and 1.

```python
#Step2 
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x:x[1])
```

```python
# Step3
rec_mov = sorted(list(enumerate(similarity[119])),reverse=True,key = lambda x:x[1])[1:6]
```

```python
# Top 5 movie names according to the 0th movie(which is avatar) in the similarity matrix is
for i in rec_mov:
    print(new_df[new_df.index == i[0]]['title'])
```

# Complete function

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0] #step1
    distances = similarity[movie_index] # locating that movie in the similarity matrix
    movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])[1:6]# Recommending the top 5 Movies
    
    # movies list contains the tuple recommended movies whose 1 index 0th index contains the index of the recommended movie
    
    # we need print the name of the movie using the indes present in the movies_list tuple.
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
